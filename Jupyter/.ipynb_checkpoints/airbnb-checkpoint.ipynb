```python
nlp=spacy.load('en_core_web_md')
```

```python
similarities={}
for target in Lmoins50:
    simi = [(w, nlp(target).similarity(nlp(w))) for w in Lplus50]
    similarities[target]=sorted(simi, key=lambda x : x[1], reverse=True)[0]
```

```python
for k, v in similarities.items():
    print(f"{k:<20}{v[0]:<20}{v[1]:4f}")
```

```python
data.property_type.apply(lambda x : similarities[x][0] if (x in similarities and similarities[x][1]>0.5) else x)
```

```python
outlier = {}
for type_pro in data.property_type.unique():
    Q3 = data[data.property_type==type_pro]['price'].quantile(0.75)
    Q1 = data[data.property_type==type_pro]['price'].quantile(0.25)
    outlier[type_pro] = (Q3 + 3*(Q3-Q1))
outlier
```
